```python
#defining all imports
#ML related python imports 
import numpy as np
import pandas as pd
from pandas.plotting import scatter_matrix
import matplotlib.pyplot as plt
import matplotlib
from sklearn.model_selection import train_test_split, StratifiedShuffleSplit
#data retrieval related imports
from urllib import request
import os
import tarfile
from pylab import *
#setting up universal variables
DOWNLOAD_ROOT = "https://raw.githubusercontent.com/ageron/handson-ml/master/"
HOUSING_PATH = os.path.join("datasets", "housing")
HOUSING_URL = DOWNLOAD_ROOT + "datasets/housing/housing.tgz"
```

Data retrieval related methods and making inital dataframe

```python
#imports cali housing cost data sets
def data_retrieve():
    if not os.path.isdir(HOUSING_PATH):
        os.makedirs(HOUSING_PATH)
    tgz_path = os.path.join(HOUSING_PATH, "housing.tgz")
    request.urlretrieve(HOUSING_URL, tgz_path)
    housing_tgz = tarfile.open(tgz_path)
    housing_tgz.extractall(path=HOUSING_PATH)
    housing_tgz.close()

data_retrieve()
    
#initializes master dataframe
housing_df = pd.read_csv(os.path.join(HOUSING_PATH, "housing.csv"))
```

Initial statistics and histogram for the data can be viewed 

```python
print(housing_df.describe())

%matplotlib inline
housing_df.hist(bins=50, figsize=(20, 15))
plt.show()
```

```python
#train test split pretty self explainitory
#training_housing_df, testing_housing_df = train_test_split(housing_df, test_size=.2, 
#                                                           random_state=1)

def get_train_test(): 
    housing_df["income_cat"] = np.ceil(housing_df["median_income"] / 1.5)
    housing_df['income_cat'].where(housing_df["income_cat"] > 5, 5.0, inplace=True)
    split = StratifiedShuffleSplit(n_splits=1, test_size=.2, random_state=1)
    testing_set = pd.DataFrame
    training_set = pd.DataFrame
    for train_index, test_index in split.split(housing_df, housing_df["income_cat"]):
        training_set = housing_df.loc[train_index]
        testing_set = housing_df.loc[test_index]
    for set_ in (training_set, testing_set):
        set_.drop("income_cat", inplace=True, axis=1)
    return (training_set, testing_set)
        
# .- alpha -... Bravo -.-. Charlie -.. Delta . Echo .-.. Foxtrot --. Golf .... Hotel 
# .. Idigo  Juliet -.- Kilo ..-. Lemur -- Mike -. November --- Oscar .--. Papa -.--
#Quebec -..- Romeo .-. Sierra
```

graphing the training set in a few different ways to look for patterns


```python
training_set, testing_set = get_train_test()
copy_train = training_set.copy()

#basic scatter plot to show dispersement
copy_train.plot(kind="scatter", x="longitude", y="latitude")
```

```python
#scatter plot with alpha set to .1 to visualize density
copy_train.plot(kind="scatter", x="longitude", y="latitude", alpha=.1)
```

```python
#scatter plot using cmap and radius to beter visualize price in terms of population
copy_train.plot(kind="scatter", x="longitude", y="latitude", alpha=.2, 
                s=copy_train["population"]/50, label="population", figsize=(15, 10), 
                c="median_house_value", cmap=plt.get_cmap("jet"))

plt.legend()
```

```python
#correlation matrix and scatter mattrix the plt close and plt.show simply clears and stops the corr matrix from displaying early

plt.close()
corr_matrix = copy_train.corr()
print(corr_matrix["median_house_value"].sort_values(ascending=False).astype(float))

attributes = ["median_house_value", "median_income", "total_rooms", 
              "housing_median_age"]
scatter_matrix(copy_train[attributes], figsize=(10, 7))
plt.show()
```

```python
#making new attributes for dataframe added in a few extra I felt like had more potential

copy_train["rooms_per_household"] = copy_train["total_rooms"]/copy_train['households']
copy_train["bedrooms_room"] = copy_train["total_bedrooms"]/copy_train["total_rooms"]
copy_train["population_per_household"] = copy_train["population"]/copy_train["households"]
copy_train["household_per_population"] = copy_train["households"]/copy_train["population"]
copy_train["RPPPHH"] = copy_train["total_rooms"]/copy_train["population_per_household"]
copy_train["BRPPPHH"] = copy_train["bedrooms_room"]/copy_train["population_per_household"]

#running a new correlation matrix
corr_matrix = copy_train.corr()
print(corr_matrix["median_house_value"].sort_values(ascending=False))
```

It is well worth noting here that I defined two metrics that out performed any of the metrics discussed in the book population per household seemed like an odd metric because it doesnt really define anything (opinion), household per population on the other hand can act as a measure of supply and demand and a measure of how robus the job market of an area is. Below I am just plotting to look at how these features correlate to median house value. I used a low alpha level so I can better visualize how these correlations are being driven by the data while ignoring some of the noise.

```python
copy_train.plot(kind="scatter", x="median_house_value", y="household_per_population", alpha=.05)
copy_train.plot(kind="scatter", x="median_house_value", y="RPPPHH", alpha=.05)
```
